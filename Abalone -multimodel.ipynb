```python
import pandas as pd
from autogluon.tabular import TabularPredictor
```

```python
train = pd.read_csv('train.csv', parse_dates=['id'])
train.head()
```

```python
train.describe()
```

```python
train.isnull().sum()
```

```python
test = pd.read_csv('test.csv', parse_dates=['id'])
test.head()
```

```python
submission = pd.read_csv("sample_submission.csv", parse_dates=['id'])
submission.head()
```

```python
from autogluon.core.metrics import make_scorer
from autogluon.tabular import TabularDataset
```

```python
ag_mean_squared_error_scorer = make_scorer(name='root_mean_squared_log_error',
                                           score_func = sklearn.metrics.root_mean_squared_log_error,
                                           optimum=0,
                                           greater_is_better=False)
```

```python
train_data = TabularDataset('/content/train.csv')  # can be local CSV file as well, returns Pandas DataFrame
label = 'Rings'  # specifies which column we want to predict
train_data = train
```

```python
predictor = TabularPredictor(label=label, problem_type= 'regression').fit(train_data=train, hyperparameters='multimodal', presets='best_quality')
```

```python
predictor.fit_summary()
```

```python
predictor.leaderboard()
```

```python
predictions = predictor.predict(test)
predictions.head()
```

```python
predictions.describe()
```

```python
sum_of_neg_val = (predictions<0).sum()
print(sum_of_neg_val)
```

```python
predictions = predictions.mask(predictions < 0, 0)
print((predictions < 0).sum()) #verify
```

```python
submission["Rings"] = predictions
submission.to_csv("submission.csv", index=False)
```
